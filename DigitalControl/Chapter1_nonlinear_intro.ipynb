# Introduction to nonlinear systems


- Most systems are nonlinear: dynamics is described by nonlinear differential equations, so you have to deal with the nonlinearities
- Linear approximation is fine for some important classes of systems: if you want design methods that are well known and robust, and don't need to get very high performances

One type of nonlinearities arise when modeling the system.

It can be easily seen that the static gain is different for different inputs. Mathematically, this is also easily proved by the differential equation. Indeed, when the transient ends, the derivatives are all equal to 0:

\begin{align}
  & 0 + |v|v = u \\
  \Leftrightarrow\: & v = \sqrt{u}
\end{align}

The other type of nonlinearities is what would be called "static nonlinearity", such as:

- relay
- dead zone
- saturation
- hysteresis
- quantization

Tools used in linear control such as Bode Plot, pole placement etc... are no more usable for nonlinear systems because the mathematical tools are based on linear differential equations.

If we really have to design a nonlinear control system, the physics will be our best option. Good physical understanding can ease the design of a controller.

Here are a list of common nonlinear behaviours:

- Multiple equilibria
- Finite escape time
- Limit cycles
- Chaos
